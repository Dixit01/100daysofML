## 1) Working with Pandas Series

```python
import pandas as pd
```

------


    1.1) Creating Series Object From List

```python
Items = ['Icecream', 'Burger', 'Samosa', 'Soft Drink']
pd.Series(Items)
```

```python
#Boolean
Boolean = [True, False, False, True, False]
pd.Series(Boolean)
```

    1.2) Create Series Object from Dictionary

```python
Dict_Series = {
    'Apple': 'Fruit', 
    'Blue': 'Colour',
    'Husky': 'Dog',
    'Shark':'Fish'
}

pd.Series(Dict_Series)
```

    1.3) Parameters and Arguments

```python
fruits_names = ['Orange', 'Apple', 'Grape', 'Blueberry', 'Plum','Mango']
days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thrusday', 'Friday','Saturday']

print(pd.Series(fruits_names, days_of_week))
```

```python
print(pd.Series(data=fruits_names, index=days_of_week))
```

```python
pd.Series(fruits_names, index=days_of_week)
```

```python
# using Custom Function

def classify_performance(number):
    if number < 5000:
        return 'Normal'
    elif number >= 5000 and number < 10000:
        return 'Good'
    else:
        return 'Incredible!'
```

```python
fortune_company_11.apply(classify_performance)
```

```python
fortune_company_11.apply(classify_performance).value_counts()
```

```python
# prefix $ dollar sign
fortune_company_11.apply(lambda Revenue: '$'+str(Revenue))
```

```python
#map() will look at the index of the second series and try to map accordingly 
    #and return the values of second series.

fortune_company_11 = pd.read_csv('Pandas_Data/fortune_1000.csv',usecols = ['Company'],squeeze = True)
fortune_company_11_t = pd.read_csv('Pandas_Data/fortune_1000.csv',usecols = ['Company', "Profits"])
print(fortune_company_11)
fortune_company_11_type = fortune_company_11_t.set_index('Company').squeeze().drop_duplicates()
print(fortune_company_11_type)
fortune_company_11.map(fortune_company_11_type)
```

```python
# In case of mistmatch index, return NAN
fortune_company_11.values[0] = "ZZZZ"

fortune_company_11.map(fortune_company_11_type)
```
