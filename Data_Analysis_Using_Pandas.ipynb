```python
emp = pd.read_csv("Pandas_Data/employees.csv")
emp
```

    Memory Optimisation

```python
emp.info()
```

In emp dataframe
-> Some columns are having missing data

-> Start Date and Last Login Time are imported as Object(string) type. This will prevent us 
    from using DateTime related methods in the future. So we need to convert them to correct datatype.

-> Senior Management are True/False value, Team and Gender are in object, We can use category data type


```python
# pd.to_datetime() Method
emp["Start Date"] = pd.to_datetime(emp['Start Date'])
emp["Start Date"]
```

```python
# For Last Login Time Column, we only have time without date. So when converting to datetime,
# it gonna use today date by default.

emp['Last Login Time'] = pd.to_datetime(emp['Last Login Time'])
emp['Last Login Time'].head(3)
```

```python
#Converting Senior Managment column to Boolean data type
emp['Senior Management'] = emp['Senior Management'].astype('bool')
emp['Senior Management']
```

```python
# to categorical data type
emp['Gender'] = emp['Gender'].astype('category')
emp.head(3)
```

```python
emp.info()
```

memory usage reduced from 64+ KB to 49+ KB

```python
emp['Team'] = emp['Team'].astype('category')
emp.info()
```

```python
emp_1 = pd.read_csv('Pandas_Data/employees.csv', parse_dates=['Start Date', 'Last Login Time'])

emp_1['Senior Management'] = emp_1['Senior Management'].astype('bool')
emp_1['Gender'] = emp_1['Gender'].astype('category')
emp_1['Team'] = emp_1['Team'].astype('category')
emp_1.info()
```
